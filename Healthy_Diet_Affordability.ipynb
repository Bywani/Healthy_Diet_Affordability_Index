## Data Extraction - PostgresSQL Database

```python
# Establishing a connection to the database.
engine = create_engine('postgresql://postgres:Thea1204%40@localhost:5432/dataengineering')
```

```python
# Defining the SQL Query that will extract the data from the database.
query1 = "SELECT * FROM healthy_diet_affordability_1"
query2 = "SELECT * FROM healthy_diet_affordability_2"
```

```python
# Fetch all the data needed from the database and store in a pandas dataframe.
df1 = pd.read_sql(query1, engine)
df2 = pd.read_sql(query2, engine)
```
